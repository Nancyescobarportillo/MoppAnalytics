```python
import pandas as pd
```

```python
df = pd.read_excel("Mopp_Analytics_Exercise.xlsx")

# Display the data frame, abbreviate dataframe to 'df'
# Set it to display 20 for better overview
df.head(20)
```

```python
# Check for duplicates based on all columns
duplicates = df[df.duplicated()]

# Print the duplicate rows
print("Duplicate rows:")
print(duplicates)
```

```python
# Check for duplicates in all columns
df = df.drop_duplicates()
df
```

```python
# Looking for all Null values in each column to identify which column is needed
df.isna().sum()
```

```python
# Drop columns that will not be used
df = df.drop(columns=["NPS_handy_rating","provider_rating","when_cancelled","coupon_amt","customer_rating"])
df
```

```python
# Convert the "hourly_charge" column from pennies to dollars
df["hourly_charge"] = df.hourly_charge / 100
df["hourly_charge"].head
```

```python
 # Test for null values
assert df.isna().sum().sum() == 0

# Test for duplicates
assert df.duplicated().sum() == 0
```

```python
df.to_excel('Mopp_Analytics_Exercise_Clean.xlsx', index=False)
```
